```python
# use the audio file as the audio source
r = sr.Recognizer()
with sr.AudioFile(filename) as source:
    audio_data = r.record(source)  # read the entire audio file
    text = r.recognize_google(audio_data)
    print (text)
```

```python
with open('meeting-words.txt' , 'w')as f:
  f.writelines(text)
```
